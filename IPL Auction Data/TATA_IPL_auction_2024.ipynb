


# Web Scrapping of IPL Auction 2024




## Import Libraries

```python
import requests
import pandas as pd
from bs4 import BeautifulSoup
```

```python
url = "https://www.iplt20.com/auction/2024"
r = requests.get(url)
print(r)
```

```python
soup = BeautifulSoup(r.text,"lxml")
table = soup.find("table",class_="ih-td-tab auction-tbl")
```

```python
table
```

```python
header = table.find_all("th")
header
```

```python
titles = []
for i in header:
    title = i.text
    titles.append(title)
print(titles)
```

```python
#create a dataframe
df = pd.DataFrame(columns = titles)
df
```

```python
rows = table.find_all("tr")
for i in rows[1:]:
    data = i.find_all("td")
    row = [tr.text for tr in data]
    l = len(df)
    df.loc[l] = row
df
```

```python
df['TEAM'] = df['TEAM'].str.replace("\n", " ")

```

```python
df['FUNDS REMAINING'] = df['FUNDS REMAINING'].str.replace("₹", " ")
```
